```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
station_zillow = pd.read_csv('https://raw.githubusercontent.com/arnenewman11/project-one/master/StationNeighborhoodPriceAction.csv')
station_zillow.head()
```

```python
stations = station_zillow['RegionName'].value_counts()
stations.index
```

```python
#bar chart that shows the number of bike stations for each neighborhood
stations.plot.bar(figsize=(15,6))
plt.title('Number of bike share stations in each Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Number of bike share stations')
plt.show()
```

```python
#lists with average price change for each neighborhood over each 3, 6 and 12 month period
three = []
six = []
twelve = []
for index, row in station_zillow.iterrows():
    change3 = ((row['3']-row['-3'])/row['-3'])*100
    if (np.isnan(change3) != True):
        three.append(change3)
    change6 = ((row['6']-row['-6'])/row['-6'])*100
    if (np.isnan(change6) != True):
        six.append(change6)
    change12 = ((row['12']-row['-12'])/row['-12'])*100
    if (np.isnan(change12) != True):
        twelve.append(change12)
three
```

```python
#total average change across each period for each time period
three_month_change = np.mean(three)
six_month_change = np.mean(six)
twelve_month_change = np.mean(twelve)
```
